```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
color = ["red","green","green","blue","red"]

df = pd.DataFrame(color, columns=["color"])
df 
```

```python
lbl_enc = LabelEncoder()
df["lbl_encoded"] = lbl_enc.fit_transform(df[["color"]])
df
```

```python
#ordinal encoding

data = ["small","medium","large","small","medium","large"]
of = pd.DataFrame(data,columns=["size"])

of
```

```python
from sklearn.preprocessing import OrdinalEncoder

OrEncoder = OrdinalEncoder(categories=[["small","medium","large"]])

of["or_encoded"]= OrEncoder.fit_transform(of[["size"]])
of
```

```python
# target Ordinal encoder
data={
    "city":["chennai","madurai","coimbatore","chennai","madurai","coimbatore"],
    "price":[200,100,300,400,200,500]
}

df_city = pd.DataFrame(data)
df_city
```

```python
mapdata = df_city.groupby(["city"])["price"].mean().to_dict()
print(mapdata)
df_city["city_encoded"] = df_city["city"].map(mapdata)
df_city
```
