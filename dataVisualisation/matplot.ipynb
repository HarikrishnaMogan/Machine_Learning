```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,5,7,9,10]

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Basic Graph plot")
plt.show()
```

```python
plt.plot(x,y,color="red",linestyle="--", marker="o",linewidth=2, markersize="9")
plt.grid(True)
```

```python
y1=[1,4,7,9,10]
y2=[6,9,12,34,50]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x, y1, color="green")
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y2,color="red")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,y2, color="blue")
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,y1, color="blue")
plt.title("Plot 4")
```

```python
cat=["A","B","C","D","E"]
values=[5,6,3,8,9]

plt.bar(cat,values,color="blue")
```

```python
data = [1,2,3,4,5,6,7,7,7,8,8,8,9,9,9]

plt.hist(data,bins=5, color="orange", edgecolor="black")
```

```python
plt.scatter(x,y ,color="blue",marker="*")
```

```python
lab=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)

plt.pie(sizes,labels=lab,colors=colors,autopct="%1.1f%%",explode=explode,shadow=True)
```
